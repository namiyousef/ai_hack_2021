# quick_exploration_housing

Exploration notebook to quicklt generate profile reports to streamline work

```python
# libs

import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# quickly explore the housing dataset
 
root_path = '../data/Boston_Housing_Market_Challenge/{}'

file_names = ['boston.csv', 'boston_corrected.csv']
```

```python
dfs = [pd.read_csv(root_path.format(name)) for name in file_names]
```

```python
dfs[0]
```

```python
dfs[1]
```

```python
dfs[1].groupby(by = 'TOWN').mean()
```

```python
profiles = [
    ProfileReport(df, title=name[:-4], explorative=True) for df, name in zip(dfs, file_names)
]
```

```python
for profile, name in zip(profiles, file_names):
    profile.to_file(name[:-3]+ 'html')
```
