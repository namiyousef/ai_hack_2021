```python
import pandas as pd
import folium
```

```python
root_path = '../data/Boston_Housing_Market_Challenge/'
dataset_paths = ['boston_corrected.csv', 'boston_fixed.csv']
```

```python
dfs = [pd.read_csv(root_path + path) for path in dataset_paths]
```

```python
dfs[0].head()
```

```python
def plot_folium(df):
    m = folium.Map([42.3557, -71.0572],zoom_start=10, tiles="cartodbpositron",)
    for lat, lon, town in zip(df['LAT'], df['LON'], df.TOWN): 
        if town == "Nahant":
            folium.CircleMarker([lat, lon],
                      radius=5,
                  #fill=False,
                        fill_opacity=0.8,
                        fill_color = 'yellow',
                  color = 'black',
                        weight = 0.3
                              ).add_to(m)
        else:
            folium.CircleMarker([lat, lon],
                      radius=5,
                  #fill=False,
                        fill_opacity=0.8,
                        fill_color = 'blue',
                  color = 'black',
                        weight = 0.3
                              ).add_to(m)
    return m
```

```python
plot_folium(dfs[0])
```

```python
plot_folium(dfs[1])
```
